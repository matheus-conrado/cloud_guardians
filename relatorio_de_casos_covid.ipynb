```python
from ETL.help_function.connect_rds_mysql import connect_mysql
import pandas as pd
```

```python
db = connect_mysql()
```

```python

query_1 = """ 
        with 
        Aux as (
        select 
            cc.country_name
            ,max(cc.confirmed) as confirmed
        from 
            Cases_covid cc
        group by
            country_name
        order by 
            confirmed desc
        limit 1)
        select  
                cc2.country_name as "Pais"
               ,cc2.confirmed as "Número de Casos Confirmados"
               ,cc2.date_register as "Data"
        from 
            Cases_covid cc2 
        inner join 
            Aux ax on ax.country_name = cc2.country_name 
        limit 500"""

query_2 = """ 
        with 
        Aux as (
        select 
            cc.country_name
            ,max(cc.deaths) as deaths
        from 
            Cases_covid cc
        group by
            country_name
        order by 
            deaths desc
        limit 1)
        select  
                cc2.country_name as "Pais"
               ,cc2.deaths as "Número de Mortos"
               ,cc2.date_register as "Data"
        from 
            Cases_covid cc2 
        inner join 
            Aux ax on ax.country_name = cc2.country_name 
        limit 500"""

query_3 = """
        select
             country_name as "País"
            ,max(deaths) as deaths
        from 
            Cases_covid cc 
        group by 
            country_name
        order BY 
            deaths DESC
        limit 10 """

query_4 = """
        select 
             country_name as "País"
            ,max(confirmed) as confirmed
        from 
            Cases_covid cc 
        group by 
            country_name
        order BY 
            confirmed DESC
        limit 10"""
```

```python
df1 = pd.read_sql(query_1,db)
df2 = pd.read_sql(query_2,db)
df3 = pd.read_sql(query_3,db)
df4 = pd.read_sql(query_4,db)
```
